###2.Data Manipulation:
a.Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes. [Hint: Use sum method with appropriate axis] b.Create a new row -“totals” that is the sum of all inmates benefitted through each mode across all states.

```python
df["total_benefitted"] = df.iloc[:,2:].sum(axis=1)
df
```

```python
totals = df.iloc[:, 2:].sum()

totals.tolist()

df.loc["totals"] = ["Total"]+["2013"] + totals.tolist()
df
```

####3.Plotting:
a.Make a bar plot with each state name on the x -axis and their total benefitted inmates astheir bar heights. Which state has the maximum number of beneficiaries? [Hint: Use bar method of pyplot] b.Make a pie chart that depicts the ratio among different modes of benefits.[Hint: Use pie method of pyplot]

```python
import matplotlib.pyplot as plt
```

```python
df_n= df['STATE/UT'][:-1]
df_n
```

```python
plt.bar(df['STATE/UT'][:-1],df['total_benefitted'][:-1])
plt.xlabel("STATE/UT")
plt.ylabel("Total Benefitted Inmates")
plt.title("Total Benefitted Inmates by State")
plt.show()

max_beneficiaries_state = df['STATE/UT'][df['total_benefitted'][:-1].idxmax()]
print(f"The state with the maximum number of beneficiaries is {max_beneficiaries_state}.")
```

```python
labels = df.columns[2:-1]  # Exclude non-numeric columns and the 'total_benefitted' column
values = df.iloc[-1].loc[labels]
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Inmates by Mode of Benefit")
plt.show()
```
