```python
import numpy as np
import pandas as pd
df_mathscore = pd.read_csv("https://raw.githubusercontent.com/tcsanitesh/DataScience/main/MathScoreTerm1.csv")
df_dssscore = pd.read_csv("https://raw.githubusercontent.com/tcsanitesh/DataScience/main/DSScoreTerm1.csv")
df_physicsscore = pd.read_csv("https://raw.githubusercontent.com/tcsanitesh/DataScience/main/PhysicsScoreTerm1.csv")

print(df_mathscore)
print(df_dssscore)
print(df_physicsscore)
```

```python
df_mathscore=df_mathscore.drop(["Name","Ethinicity"], axis=1)
df_dssscore=df_dssscore.drop(["Name","Ethinicity"], axis=1)
df_physicsscore=df_physicsscore.drop(["Name","Ethinicity"], axis=1)

print(df_mathscore)
print(df_dssscore)
print(df_physicsscore)
```

```python
count_df_mathscore = df_mathscore["Score"].isna().sum()
count_df_dssscore = df_dssscore["Score"].isna().sum()
count_df_physicsscore = df_physicsscore["Score"].isna().sum()

print(count_df_mathscore)
print(count_df_dssscore)
print(count_df_physicsscore)
```

```python
df_mathscore["Score"]=df_mathscore["Score"].fillna(0)
count_df_mathscore = df_mathscore["Score"].isna().sum()
print(count_df_mathscore)
```

```python
df_dssscore["Score"]=df_dssscore["Score"].fillna(0)
count_df_dssscore = df_dssscore["Score"].isna().sum()
print(count_df_dssscore)
df_physicsscore["Score"]=df_physicsscore["Score"].fillna(0)
count_df_physicsscore = df_physicsscore["Score"].isna().sum()
print(count_df_physicsscore)
```

```python
df = [df_mathscore,df_dssscore,df_physicsscore] # list of all df
df
```

```python
merged_df = pd.concat(df,ignore_index=True)
merged_df
```

```python
merged_df['Sex'] = merged_df['Sex'].replace({"M": 1, "F": 2})
merged_df
```

```python
merged_df.to_csv("ScoreFinal.csv")
```

```python
import numpy as np
import pandas as pd
df_mathscore = pd.read_csv("https://raw.githubusercontent.com/tcsanitesh/DataScience/main/MathScoreTerm1.csv")
df_dssscore = pd.read_csv("https://raw.githubusercontent.com/tcsanitesh/DataScience/main/DSScoreTerm1.csv")
df_physicsscore = pd.read_csv("https://raw.githubusercontent.com/tcsanitesh/DataScience/main/PhysicsScoreTerm1.csv")

print(df_mathscore)
print(df_dssscore)
print(df_physicsscore)
```

```python
avg_math=df_mathscore["Score"].mean()
avg_dss=df_dssscore["Score"].mean()
avg_phy=df_physicsscore["Score"].mean()
print(avg_math)
print(avg_dss)
print(avg_phy)
```

```python
df_mathscore["Score"]=df_mathscore["Score"].fillna(avg_math)
count_df_mathscore = df_mathscore["Score"].isna().sum()
print(count_df_mathscore)
df_dssscore["Score"]=df_dssscore["Score"].fillna(avg_dss)
count_df_dssscore = df_dssscore["Score"].isna().sum()
print(count_df_dssscore)
df_physicsscore["Score"]=df_physicsscore["Score"].fillna(avg_phy)
count_df_physicsscore = df_physicsscore["Score"].isna().sum()
print(count_df_physicsscore)
```

```python
df = [df_mathscore,df_dssscore,df_physicsscore]
```

```python
merged_df = pd.concat(df,ignore_index=True)
merged_df
```

```python
list_eth = merged_df["Ethinicity"].unique()
#df_dssscore = pd.read_csv("https://raw.githubusercontent.com/tcsanitesh/DataScience/main/DSScoreTerm1.csv")
#df_physicsscore = pd.read_csv("https://raw.githubusercontent.com/tcsanitesh/DataScience/main/PhysicsScoreTerm1.csv")

print(list_eth)
#print(df_dssscore)
#print(df_physicsscore)
```

```python
ethnicity_mapping = {
    'White American': 1,
    'European American': 2,
    'Hispanic': 3,
    'African American': 4,
    'Other': 5,
}

merged_df['Ethinicity'] = merged_df['Ethinicity'].replace(ethnicity_mapping)
```

```python
merged_df.to_csv("Enhanced_file.csv")
```
